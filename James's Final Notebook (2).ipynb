For my dataset, I chose to focus on the number of fire incidents in Pittsburgh Neighborhoods. I chose this as my main topic because I felt it could be a main factor in finding what the safest neighborhood to live in was. The dataset chosen is called "Fire Incidents in Pittsburgh" and can be found on the Western Regional Data Center website. 

My first step was importing in pandas and matplot to be able to use the tables and charts that are seen below. 

```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python

fire = pd.read_csv("fi.csv", parse_dates=True)

fire.head(5)
```

Next was counting the number of times a neighborhood had appeared in the table, I did this by using the value_counts command and having it focus on the neighborhood column. 

```python
fire['neighborhood'].value_counts()
```

We have our general info of which neighborhoods have the most and least fire incidents, but there could still be more to it to better understand what were looking at, so how about we make a graph. I'm going to make a bar graph, with the x values representing the neighborhoods and the y values representing the number of fire incidents. 

```python
graph = fire['neighborhood'].value_counts()
graph.plot(kind = 'bar', xlabel = 'Neighborhoods', ylabel = 'Fire Incidents')
```

So theres seems to be an issue of having too many neighborhoods squished into too small of a space, but its a simple fix, so I added that along with some other bits of code to make it easier to read and understand. 

```python
graph = fire['neighborhood'].value_counts()

graph.plot(kind = 'bar', 
           figsize=(19,13),
           fontsize = 12)
plt.title("Number of Fire Incidents in Pittsburgh Neighborhoods", fontsize=30)
plt.xlabel("Neighborhoods", fontsize = 25)
plt.ylabel("Fire Incidents", fontsize = 25)


```

This graph upbove works, but it has too much information, so how about instead of trying to zoom in and read this, we'll just find the 5 neighborhoods with the least amount of fire incidents. 

```python
best =["Ridgemont","Mt. Oliver","East Carnegie","Regent Sq.","Mt. Oliver Borough"]
incidents =[12,12,11,9,6]
plt.subplots(figsize=(10, 5))
plt.title("Neighborhoods with the Least Amount of Fire Incidents", fontsize=20)
plt.xlabel("Neighborhoods", fontsize = 15)
plt.ylabel("Fire Incidents", fontsize = 15)
plt.bar(best, incidents)
plt.show()
```

```python
most =["Carrick","South Sides Flats","Central Business District","Squirrel Hill South", "Hazelwood"]
incidents2 =[307,290,290,269,257]
plt.subplots(figsize=(12, 5))
plt.title("Neighborhoods with the Most Amount of Fire Incidents", fontsize=20)
plt.xlabel("Neighborhoods", fontsize = 15)
plt.ylabel("Fire Incidents", fontsize = 15)
plt.bar(most, incidents2)
plt.show()
```

So, the final results for for least amount of incidents and safest is Mt. Oliver Borough, although it does repeat twice so Im pretty sure it would instead be Regent Square as the safest neighborhood. And the neighborhood with the most amount of incidents is Carrick.

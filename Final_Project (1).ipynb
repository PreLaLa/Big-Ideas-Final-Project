```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
South_Side_Flats=0
Homewood_North=0
Homewood_South=0
Central_Business_District=0
Marshall_Shadeland=0
fSouth_Side_Flats=0
fHomewood_North=0
fHomewood_South=0
fCentral_Business_District=0
fMarshall_Shadeland=0
pSouth_Side_Flats=219
pHomewood_North=141
pHomewood_South=218
pCentral_Business_District=189
pMarshall_Shadeland=150
with open('C:\\Users\\prest\\Intro-to-Jupyter\\Final-Project\\apb.csv', mode = 'r') as csvfile:
    file = csv.DictReader(csvfile)
    for row in file:
        neighborhood = row['INCIDENTNEIGHBORHOOD']
        
        if neighborhood == "South Side Flats":
            South_Side_Flats += 1
                
        if neighborhood == "Homewood North":
            Homewood_North += 1
        
        if neighborhood == "Homewood South":
            Homewood_South += 1
            
        if neighborhood == "Crawford-Roberts":
            Central_Business_District += 1
            
        if neighborhood == "Marshall-Shadeland":
            Marshall_Shadeland += 1
with open("fire.csv",mode='r') as csvfile:
    file = csv.DictReader(csvfile)
    for row in file:
        neighborhood = row['neighborhood']
        
        if neighborhood == "South Side Flats":
            fSouth_Side_Flats += 1
                
        if neighborhood == "Homewood North":
            fHomewood_North += 1
        
        if neighborhood == "Homewood South":
            fHomewood_South += 1
            
        if neighborhood == "Central Business District":
            fCentral_Business_District += 1
            
        if neighborhood == "Marshall-Shadeland":
            fMarshall_Shadeland += 1
            
x = ["South Side Flats","Homewood North","Homewood South","Central Business District","Marshall-Shadeland"]
y = [South_Side_Flats/1000,Homewood_North/1000,Homewood_South/1000,Central_Business_District/1000,Marshall_Shadeland/1000]
fy = [fSouth_Side_Flats,fHomewood_North,fHomewood_South,fCentral_Business_District,fMarshall_Shadeland]
py = [pSouth_Side_Flats,pHomewood_North,pHomewood_South,pCentral_Business_District,pMarshall_Shadeland]
oy = [y[0]+fy[0]+py[0],y[1]+fy[1]+py[1],y[2]+fy[2]+py[2],y[3]+fy[3]+py[3],y[4]+fy[4]+py[4],]
plt.plot(x, y, color='b') 
plt.plot(x, fy, color='r') 
plt.plot(x, py, color='g') 
plt.plot(x, oy, color='grey')

plt.title("Overall Graph") 
plt.xlabel("Neighborhoods")
plt.ylabel("Number of Incidents/Arrests")
plt.xticks(rotation = 35)
plt.legend(["Police Incidents(1000s)","Fire Incidents","Police Arrests","Total Results"])
plt.show()
```

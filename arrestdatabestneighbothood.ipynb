```python
import pandas as pd

arrestdata = pd.read_csv('arrestdata.csv')
arrestdata['ARRESTYEAR'] = pd.to_datetime(arrestdata['ARRESTTIME']).dt.year

crime2018 = arrestdata[arrestdata['ARRESTYEAR'] == 2018]
crime2022 = arrestdata[arrestdata['ARRESTYEAR'] == 2022]

crimenumber2018 = crime2018['INCIDENTNEIGHBORHOOD'].value_counts()
crimenumber2022 = crime2022['INCIDENTNEIGHBORHOOD'].value_counts()

crimedifference = {neighborhood: crimenumber2018.get(neighborhood, 0) - crimenumber2022.get(neighborhood, 0)
                   for neighborhood in set(crimenumber2018.index).union(crimenumber2022.index)}

crdiffsorted = dict(sorted(crimedifference.items(), key=lambda item: item[1],reverse=True))

crimedifftop5 = dict(list(crdiffsorted.items())[:5])

print(crimedifftop5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
arrestdata = pd.read_csv('arrestdata.csv')
arrestdata['ARRESTYEAR'] = pd.to_datetime(arrestdata['ARRESTTIME']).dt.year
currentyear = pd.Timestamp.now().year
southsideflats = arrestdata[(arrestdata['INCIDENTNEIGHBORHOOD'] == 'South Side Flats') & (arrestdata['ARRESTYEAR'] >= currentyear - 6)]
crimeperyear = southsideflats.groupby('ARRESTYEAR').size()
plt.plot(crimeperyear.index, crimeperyear.values)
plt.show()
```
